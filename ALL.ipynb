## Tokenization Functions

```python
def get_unigrams(df, malwareType):
  count = 0
  for row in df['Opcodes']:
    count += 1
    if count > 5:
      break
    data = row.split()
    unigramList.append((data, malwareType))
```

```python
def get_bigrams(df, malwareType):
  count = 0
  for row in df['Opcodes']:
    count += 1
    if count > 5:
      break
    data = row.split()
    bigrms = list(bigrams(data))
    bigrms = [f"{first} {second}" for first, second in bigrms]
    bigramList.append((bigrms, malwareType))
```

```python
def get_trigrams(df, malwareType):
  count = 0
  for row in df['Opcodes']:
    count += 1
    if count > 5:
      break
    data = row.split()
    trigrms = list(trigrams(data))
    trigrms = [f"{first} {second} {third}" for first, second, third in trigrms]
    trigramList.append((trigrms, malwareType))
```

```python
def get_bigramstrigrams(df, malwareType):
  count = 0
  for row in df['Opcodes']:
    count += 1
    if count > 5:
      break
    data = row.split()
    if count % 2 == 0:
      bigrms = list(bigrams(data))
      bigrms = [f"{first} {second}" for first, second in bigrms]
      bigramtrigramList.append((bigrms, malwareType))
    else:
      trigrms = list(trigrams(data))
      trigrms = [f"{first} {second} {third}" for first, second, third in trigrms]
      bigramtrigramList.append((trigrms,malwareType))
```
